```python
%matplotlib notebook
import pickle
import pandas as pd
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
#读取数据
data = pd.read_csv('./file_save/feature_exctracted.csv')

dataA = data.dropna()   #清空行
dataA = dataA.drop(26,axis=0)   #去掉异常车辆26

veh_id = pd.DataFrame(dataA.iloc[:,0])  # 提出车辆编号
veh_id.columns = ['veh_id']
veh_id = veh_id.reset_index(drop = True)
dataA = dataA.drop('num',axis=1)

dataA = dataA.reset_index(drop = True)
dataA
```

```python
# 所有变量
std = StandardScaler()# 标准化
dataA_sdd = std.fit_transform(dataA)   #标准化
dataA_sdd = pd.DataFrame(dataA_sdd,columns = dataA.columns)   #转为dataframe
kmeans = KMeans(n_clusters=2, random_state=0).fit(dataA_sdd)   #跑kmeans
label = pd.DataFrame(kmeans.labels_)    #取label
label.columns = ['label']
dataA_result = dataA.join(label)
dataA_result = dataA_result.join(veh_id)
dataA_result
```

```python
# 储存
dataA_result.to_csv('./file_save/kmeans_result.csv')
```

```python
# 精简变量
dataA = dataA.drop('point_num',axis=1)
dataA_sdd = std.fit_transform(dataA)   #标准化
dataA_sdd = pd.DataFrame(dataA_sdd,columns = dataA.columns)   #转为dataframe
kmeans = KMeans(n_clusters=2, random_state=0).fit(dataA_sdd)   #跑kmeans
label = pd.DataFrame(kmeans.labels_)    #取label
label.columns = ['label']
dataA_result = dataA.join(label)
dataA_result = dataA_result.join(veh_id)
dataA_result
```

```python
# 储存
dataA_result.to_csv('./file_save/kmeans_result_less.csv')
```
