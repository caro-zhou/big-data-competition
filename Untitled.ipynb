```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('C:\competition\新能源汽车\数据：新能源乘用车公共充电需求测算模型设计\动态表.csv')
data
```

```python
# 把同一辆车的数据放在一起并按时间顺序排列
data = data.sort_values(by=['vid','daq_time'])
```

```python
# 时间信息处理
data['time'] = data['daq_time'].apply(lambda x: str(x)[:4]+'-'+str(x)[4:6]+'-'+str(x)[6:8]+' '+str(x)[8:10]+':'+str(x)[10:12]+':'+str(x)[12:])
```

```python
data['to_datetime'] = pd.to_datetime(data['time'])
data['time_diff'] = (data['to_datetime'] - data['to_datetime'].shift(1)).dt.seconds # 与上一条记录的时间差
```

```python
data
```

```python
data = data[data['time_diff'] != 0].reset_index(drop=True)
data
```

```python
data['soc_shift_1'] = data['soc'].shift(1) # 上一条记录的soc
data = data[(data['soc'] != 0) | ((data['soc'] == 0) & (data['soc_shift_1'] != 0))].reset_index(drop=True) 
# 去除soc连续为0的记录（保留第一条，仅保留一条）
```

```python
data
```

```python
# 处理一下相邻两条记录不是同一辆车的情况
# data['different_veh'] = 0
# data['vid_shift'] = data['vid'].shift(1)
# data['different_veh'] = data.apply(lambda x: 1 if x['vid']!=x['vid_shift'] else 0, axis=1)
data['codes'], _ = pd.factorize(data['vid'])
```

```python
data['code_shift'] = data['codes'].shift(1)
data['time_diff'] = data.apply(lambda x: x['time_diff'] if x['codes']==x['code_shift'] else -1, axis=1) # 与上一条记录的时间差,-1表示新的一辆车
```

```python
# 处理SOC=0:如果time_diff为-1，则取下一个记录的soc值；否则按照时间进行插值
data['time_diff_shift'] = data['time_diff'].shift(-1) # 与下一条记录的时间间隔，-1则此纪录为一辆车的最后一条
# 删除了一些数据后，重新搞一下soc shift
data['soc_shift_-1'] = data['soc'].shift(-1) # 下一条记录的soc
data['soc_shift_1'] = data['soc'].shift(1) # 上一条记录的soc
data['soc'] = data.apply(lambda x: x['soc_shift_-1']+
                         ((x['soc_shift_1'] - x['soc_shift_-1'])/ (x['time_diff'] + x['time_diff_shift'])) * x['time_diff_shift'] 
                         if (x['soc']==0 and x['time_diff']!=-1) else (x['soc_shift_-1'] if (x['soc']==0 and x['time_diff']==-1) 
                                                                       else x['soc']), axis=1)
```

```python
data.to_csv('C:\competition\新能源汽车\数据：新能源乘用车公共充电需求测算模型设计\data_processed.csv')
```
